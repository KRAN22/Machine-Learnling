```python
import xml.etree.ElementTree as ET
import pandas as pd

customer_data = {}
root = ET.parse('sample_CustomersOrders.xml').getroot()
for elem in root.iter():
    if elem.tag!= "Root" and elem.tag!="Customers" and elem.tag!="Customer" and elem.tag!="Order" and elem.tag != "Orders":
        customer_data[elem.tag] = []
for customer in root:
    print(customer)
    for customer_attrb in customer.iter():
        if customer_attrb != "Customer" and customer_attrb != "Order" and customer_attrb != "Orders":
            print(customer_attrb)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

book_data = {'id':[]}
root = ET.parse('book.xml').getroot()
for elem in root.iter():
    if elem.tag!='catalog' and elem.tag!='book':
        book_data[elem.tag] = []

for book in root.findall('book'):
    book_data['id'].append(book.get('id'))
    for book_attrb in book_data:
        if book_attrb!='id':
            s = book.find(book_attrb).text.replace('\n',' ')
            book_data[book_attrb].append(' '.join(s.split()))
pd.DataFrame(book_data)
```
