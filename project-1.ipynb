# Handling Duplicates

```python
dataFrame.duplicated()
```

```python
droped_dataFrame = dataFrame.drop_duplicates()
droped_dataFrame
```

# Handling Missing Values

```python
droped_dataFrame.isnull().sum()
```

```python
droped_dataFrame['num_room'] = droped_dataFrame['num_room'].fillna(droped_dataFrame['num_room'].median())
```

```python
droped_dataFrame.isnull().sum()
```

# Handling Outliers

```python
for col in  droped_dataFrame.columns:
    plt.boxplot(droped_dataFrame[col])
    plt.title(col)
    plt.show()
```

```python
for col in droped_dataFrame.columns:
    p0 = droped_dataFrame[col].min()
    p100 = droped_dataFrame[col].max()
    q1 = droped_dataFrame[col].quantile(0.25)
    q2 = droped_dataFrame[col].quantile(0.5)
    q3 = droped_dataFrame[col].quantile(0.75)
    iqr=q3-q1
    lc = q1 - 1.5*iqr
    uc = q3 + 1.5*iqr
    droped_dataFrame[col]=np.where(droped_dataFrame[col] >= uc,uc,droped_dataFrame[col])
    droped_dataFrame[col]=np.where(droped_dataFrame[col] <= lc,lc,droped_dataFrame[col])
```

```python
for col in  droped_dataFrame.columns:
    plt.boxplot(droped_dataFrame[col])
    plt.title(col)
    plt.show()
```
